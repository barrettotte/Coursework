**(3)** $g(t) = \frac{16t^2-32}{t}$

$\quad=\frac{\frac{d}{dt}(16t^2-32)t - \frac{d}{dt}(t)(16t^2-32)}{t^2}\quad$(quotient rule)

$\quad\quad\frac{d}{dt}(16t^2-32) \;=\; \frac{d}{dt}(16t^2) - \frac{d}{dt}(32) \;=\; 32t-0$

$\quad=\frac{(32t)t - (16t^2-32)}{t^2} \;=\; \frac{16t^2+32}{t^2}$

$\quad \frac{d}{dx}((x^2-7)^3)$

$\quad \frac{d}{dx}(x^4+1)$

$\quad\quad \frac{d}{dx}(x^4) + \frac{d}{dx}(1) \;=\; 4x^3 + 0$

**(11)** $f(x) = (x^3-1)\sqrt{3x^2+4}$

$\frac{d}{dx}(x^3-1)\sqrt{3x^2+4} + \frac{d}{dx}(\sqrt{3x^2+4})(x^3-1) \quad$ (product rule)

$\frac{d}{dx}(x^3-1) = 3x^2$

$\frac{d}{dx}(\sqrt{3x^2+4})$

$\quad \text{let } f=\sqrt{u},\;u=3x^2+4 \quad$ (chain rule)

$\quad \frac{d}{du}(\sqrt{u})\frac{d}{dx}(3x^2+4) \;=\; \frac{1}{2\sqrt{u}} \frac{d}{dx}(3x^2+4)$

$\quad =\frac{1}{2\sqrt{3x^2+4}}\frac{d}{dx}(3x^2+4)$
$\;=\;\frac{1}{2\sqrt{3x^2+4}}\cdot 6x$
$\;=\frac{3x}{\sqrt{3x^2+4}}$

<br>

$=3x^2\sqrt{3x^2+4} + \frac{3x}{\sqrt{3x^2+4}}(x^3-1)$

$=3x^2\sqrt{3x^2+4} + \frac{3x(x^3-1)}{\sqrt{3x^2+4}}$
$\;=\;\frac{3x(x^3-1)}{\sqrt{3x^2+4}} + \frac{3x^2\sqrt{3x^2+4}\sqrt{3x^2+4}}{\sqrt{3x^2+4}}$

$\frac{3x(x^3-1)+3x^2(3x^2+4)}{\sqrt{3x^2+4}} \;=\;\frac{12x^4+12x^2-3x}{\sqrt{3x^2+4}}$

<br>


**(15)** $f(x) = \left(\frac{x}{x^3+1}\right)^{-4}$

$\text{let } f = u^{-4},\;u=\frac{x}{x^3+1} \quad \text{(chain rule)}$

$\quad \frac{d}{du}(u^{-4})\frac{d}{dx}(\frac{x}{x^3+1})$

$\quad -\frac{4}{u^5}\frac{d}{dx}\left(\frac{x}{x^3+1}\right)$
$\;=\; -\frac{4}{\left(\frac{x}{x^3+1}\right)^5}\frac{d}{dx}\left(\frac{x}{x^3+1}\right)$

$\quad \frac{d}{dx}\left(\frac{x}{x^3+1}\right)$

$\quad\quad \frac{\frac{d}{dx}(x)(x^3+1) - \frac{d}{dx}(x^3+1)x}{(x^3+1)^2}\quad$ (quotient rule)

$\quad\quad = \frac{1\cdot (x^3+1) -3x^2x}{(x^3+1)^2}$
$\;=\; \frac{-2x^3+1}{(x^3+1)^2}$

$= -\frac{4}{\left(\frac{x}{x^3+1}\right)^5}\cdot \frac{-2x^3+1}{(x^3+1)^2}$
$\;=\; -\frac{4(x^3+1)^5}{x^5} \cdot \frac{-2x^3+1}{(x^3+1)^2}$

$= -\frac{4(x^3+1)^5(-2x^3+1)}{x^5(x^3+1)^2}$
$\;=\; -\frac{4(x^3+1)^3(-2x^3+1)}{x^5}$

**(17)** $f(x) = (\sin x + \cos x)^2$

$\frac{d}{dx}(\cos^2(x) + \sin^2(x) + 2\cos(x)\sin(x))$

$\frac{d}{dx}(1+\sin(2x)) \quad$ (trig identities)

$\frac{d}{dx}(1) + \frac{d}{dx}(\sin(2x))$

$\quad \text{let } f=\sin(u),\;u=2x \quad \text{(chain rule)}$

$\quad =\frac{d}{du}(\sin(u)) = \cos(u)$

$\quad =\cos(2x)\frac{d}{dx}(2x) \;=\; 2 \cos(2x)$

<br>

**(19)** $f(t) = \tan 2t + \sec 3t$

$\frac{d}{dt}(\tan(2t)) + \frac{d}{dt}(\sec(3t))$

$\frac{d}{dt}(\tan(2t))$

$\quad \text{let } f=\tan(u),\;u=2t\quad\text{(chain rule)}$

$\quad \frac{d}{du}(\tan(u))\frac{d}{dt}(2t)$
$\;=\; \sec^2(u)\frac{d}{dt}(2t)$

$\quad =\sec^2(2t)\frac{d}{dt}(2t) \;=\; 2 \sec^2(2t)$

$\frac{d}{dt}(\sec(3t))$

$\quad \text{let } f = \sec(u),\;u=3t\quad\text{(chain rule)}$

$\quad \frac{d}{du}(\sec(u))\frac{d}{dt}(3t)$
$\;=\; \sec(u)\tan(u)\frac{d}{dt}(3t)$
$\;=\; \sec(3t)\tan(3t)\frac{d}{dt}(3t)$

$\quad = 3\sec(3t)\tan(3t)$

$= 2\sec^2(2t) + 3\sec(3t)\tan(3t)$
