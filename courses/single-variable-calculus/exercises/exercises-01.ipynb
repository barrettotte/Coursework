```python
from IPython.display import display, Math
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import sympy.plotting.plot as symplot

# generic util for plotting simple f(x) with symplot
def symplot_fx(x, fx, xlim=[-10,10], ylim=[-10,10], dsp_fx=True):
  if dsp_fx:
    display(Math('$f(x) = %s$' %sym.latex(fx)))

  p = symplot(fx, (x, ylim[0], ylim[1]), xlim=xlim, ylim=ylim, show=False)
  p.show()
```
