**(1)** Evaluate the following integrals

<br>

**(a)** $\int_0^2 \frac{xdx}{(1+x^2)^2}$

let $u=1+x^2$

$\frac{d}{dx}(1+x^2) = 2x \implies du=2xdx \quad dx=\frac{1}{2}du$

$\int\frac{x}{u^2} \cdot \frac{1}{2x}du$
$\;=\; \int\frac{1}{2u^2}du$

$u = 1+x^2 \implies 1+(0)^2=1 \quad 1+(2)^2=5 \quad\text{(adjust bounds)} $

$\int_1^5\frac{1}{2u^2}du$
$\;=\; \frac{1}{2}\int_1^5 \frac{1}{u^2}du$
$\;=\; \frac{1}{2}\int_1^5 u^{-2}du$

$\frac{1}{2}\left[ \frac{u^{-1}}{-2+1} \right]_1^5$
$\;=\; \frac{1}{2}\left[ -\frac{1}{u} \right]_1^5$
$\;=\; (-\frac{1}{5}) - (-\frac{1}{1})$
$\;=\; \frac{1}{2}\left[ -\frac{1}{5} + 1 \right]$
$\;=\; \frac{2}{5}$

**answer:** $\quad \frac{2}{5}$

<br>


**(b)** $\int_{-\pi/2}^{\pi/2} \sin^6(x)\cos(x)dx$

let $u=\sin(x)$

$\frac{d}{dx}(\sin(x)) = \cos(x) \implies du=\cos(x)dx \quad dx=\frac{1}{\cos(x)}du$

$\int u^6\cos(x)\frac{1}{\cos(x)}du$
$\;=\; \int u^6du$

$u = \sin(x) \implies \sin(-\frac{\pi}{2})=-1 \quad \sin(\frac{\pi}{2})=1$

$\int_{-1}^1 u^6du$
$\;=\; \frac{u^7}{7}\big|_{-1}^1$
$\;=\; \frac{1}{7} - (-\frac{1}{7})$
$\;=\; \frac{2}{7}$

**answer:** $\quad \frac{2}{7}$

**(2)** Find the approximations to $\int_0^{\pi/2}\cos(x)dx$

<br>

**(a)** Using the upper Riemann sum with two intervals

$\Delta x = \frac{b-a}{n} \;=\; \frac{\frac{\pi}{2}}{2} \;=\; \frac{\pi}{4}$

$\int_0^{\frac{\pi}{2}}\cos(x)dx \approx \frac{\pi}{4}\cdot [\cos(0)+\cos(\frac{\pi}{4})]$
$\;=\; \frac{\pi}{4} + \frac{\pi}{4}\cos(\frac{\pi}{4})$
$\;=\; \frac{\pi}{4} \left( 1+\frac{\sqrt{2}}{2} \right)$

**answer:** $\quad \frac{\pi}{4} \left( 1+\frac{\sqrt{2}}{2} \right)$

<br>

**(b)** Using the trapezoidal rule with two intervals

using prior problem: $\Delta x = \frac{\pi}{4}$

$\int_0^{\frac{\pi}{2}}\cos(x)dx \approx \frac{\frac{\pi}{4}}{2}\cdot \left[ \cos(0) + \cos(\frac{\pi}{4}) \right] + \frac{1}{2}\frac{\pi}{4}\frac{\sqrt{2}}{2}$
$\;=\; \frac{\pi(\sqrt{2}+1)}{8}$

**answer:** $\quad \frac{\pi(\sqrt{2}+1)}{8}$

<br>

**(c)** Using Simpson's rule with two intervals

using prior problem: $\Delta x = \frac{\pi}{4}$

$\int_0^{\frac{\pi}{2}} \approx \frac{\frac{\pi}{4}}{3}\cdot [\cos(0) + 4\cos(\frac{\pi}{4})+\cos(\frac{\pi}{2})]$
$\;=\; \frac{\pi}{12}[1 + 2\frac{\sqrt{2}}{2}]$
$\;=\; \frac{\pi}{12}(1 + 2\sqrt{2})$

**answer:** $\quad \frac{\pi}{12}(1 + 2\sqrt{2})$

**(3)** Find the volume of the solid of revolution formed by revolving the y-axis the region enclosed by $y=\cos(x^2)$ and the x-axis (central hump, only)


$ V = \int_a^b 2\pi xf(x)dx \quad\text{(shell method)}$

$ V = \int\limits_0^{\sqrt{\frac{\pi}{2}}} 2\pi x\cdot \cos(x^2)dx$
$ \;=\; 2\pi\int\limits_0^{\sqrt{\frac{\pi}{2}}} x\cos(x^2)dx$

let $u=x^2$

$\frac{d}{dx}(x^2) = 2x \implies du=2xdx \quad dx=\frac{1}{2x}du$

$2\pi \int x\cos(u)\frac{1}{2x}du$
$\;=\; 2\pi\int \frac{\cos(u)}{2}du$

$u = x^2 \implies (0)^2=0 \quad (\sqrt{\frac{\pi}{2}})^2 = \frac{\pi}{2}$

$2\pi\frac{1}{2} \int_0^{\frac{\pi}{2}}\cos(u)du$

$\pi\left[\sin(u)\right]_0^\frac{\pi}{2}$
$\;=\; \pi[\sin(\frac{\pi}{2}) - \sin(0)]$
$\;=\; \pi[1-0]$
$\;=\; \pi$

**answer:** $\quad \pi$
