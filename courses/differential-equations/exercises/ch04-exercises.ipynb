## 4.2 - Reduction of Order

**1.** Use reduction of order or formula (5) to find a second solution $y_2(x)$ given a solution of the given differential equation $y_1(x)$.

- $y''-4y'+4y=0$
- $y_1=e^{2x}$

$y''+p(x)y'+q(x)y=0$
$\implies p(x)=-4,\quad q(x)=4$

$y_1(x)=e^{2x}$

$y_2(x)=y_1(x)\int\frac{e^{-\int p(x)dx}}{(y_1)^2}dx$

$=e^{2x}\int\frac{e^{-\int -4dx}}{(e^{2x})^2}dx$

$=e^{2x}\int\frac{e^4x}{e^4x}dx$

$=e^{2x}x$

<hr>

**3.** Use reduction of order or formula (5) to find a second solution $y_2(x)$ given a solution of the given differential equation $y_1(x)$.

- $y''+16y=0$
- $y_1=\cos(4x)$

$y''+p(x)y'+q(x)y=0$
$\implies p(x)=0,\quad q(x)=16$

$y_1=\cos(4x)$

$y_2(x)=y_1(x)\int\frac{e^{-\int p(x)dx}}{(y_1)^2}dx$

$=\cos(4x)\int\frac{e^{-\int 0dx}}{(\cos(4x))^2}dx$

$=\cos(4x)\int\frac{k}{\cos^2(4x)}dx$

$=\cos(4x)\int k\sec^2 4xdx$

$=\cos(4x)(k)\cdot\frac{1}{4}\tan4x$

$=\frac{k}{4}\sin4x$

$=c_2\sin 4x = \sin 4x$

## 4.6 - Variation of Parameters

**1.** Solve the differential equation by variation of parameters. $\quad y''+y=\sec(x)$

$y''+p(x)y'+q(x)y=f(x)$

$y=e^{mx},\quad y'=me^{mx},\quad y''=m^2e^{mx}$

$m^2e^{mx}+e^{mx}=0$

$e^{mx}(m^2+1)=0$

$m^2+1=0\implies m_1=m_2=\pm i$

$y_h=c_1y_1+c_2y_2$

$=c_1\cos x+c_2\sin x$

<br>

$W(y_1,y_2)=$
$\begin{vmatrix}\cos x&\sin x\\-\sin x&\cos x\end{vmatrix}$
$=\cos^2x-(-\sin^2x)$
$=1$

$W_1=\begin{vmatrix}0&y_2\\f(x)&y'_2\end{vmatrix}$
$=\begin{vmatrix}0&\sin x\\\sec x&\cos x\end{vmatrix}$
$=0-(\sin x)(\sec x)$

$=-\tan x$

$W_2=\begin{vmatrix}y_1&0\\y'_1&f(x)\end{vmatrix}$
$=\begin{vmatrix}\cos x&0\\-\sin x&\sec x\end{vmatrix}$
$=(\cos x)(\sec x)-0$
$=1$

<br>

$u'_1=\frac{W_1}{W(y_1,y_2)}$
$=\frac{-\tan x}{1}=-\tan x$

$u'_2=\frac{W_2}{W(y_1,y_2)}$
$=\frac{1}{1}=1$

$u_1=\int u'_1dx$
$=\int -\tan x dx$
$=\int -\frac{\sin x}{\cos x}dx$
$=\ln(\cos x)$

$u_2=\int u'_2dx$
$=\int 1dx = x$

<br>

$y_p=u_1y_1+u_2y_2$
$=\ln(\cos x)(\cos x)+x\sin x$

$y=y_h+y_p$

$y = c_1\cos x+c_2\sin x+\cos x\ln(\cos x)+x\sin x$

$x^2\frac{d^2y}{dx^2}-2y=0$

<br>

$m_1=-1,\quad m_2=2$

<br>

$y=c_1x^{-1}+c_2x^2$
