**3.** The population of a town grows at a rate proportional to the population present at time $t$. The initial population of 500 increases by 15% in 10 years. What will be the population in 30 years? How fast is the population growing at $t=30$?

$\frac{dP}{dt}=kP$

$P(0)=500\quad P(10)=1.15(500)=575$

$\frac{dP}{P}=kdt$

$\int\frac{1}{P}dp=k\int1dt$

$\ln(P)=kt+c$

$e^{\ln(P)}=e^{(kt+c)}$

$P=e^{kt}e^{c}=ce^{kt}$

<br>

$500=ce^0\implies c=500$

$P=500e^{kt}$

<br>

$575=400e^{10k}$

$\frac{575}{500}=e^{10k}$

$\ln(1.15)=10k$

$k=\frac{\ln(1.15)}{10}=\frac{0.14}{10}=0.014$

<br>

$P=500e^{0.014t}$

$P=500e^{0.014(30)}\approx760\text{ people}$

$\frac{dP}{dt}=0.014(760)\approx 11\text{ people/year}$

**5.** The radioactive isotope of lead, $\text{Pb-}209$, decays at a rate proportional to the amount present at
time $t$ and has a half-life of 3.3 hours. If 1 gram of this isotope is present initially, how long will it take for 90% of the lead to decay?

$\frac{dx}{dt}=-kx$

$t(1)=0\quad t(0.5)=3.3$

$\frac{dx}{x}=-kdt$

$\int\frac{1}{x}dx=-k\int1dt$

$\ln(x)=-kt+c$

$e^{ln(x)}=e^{(-kt+c)}$

$x=e^{-kt}e^{c}=ce^{-kt}$

$1=ce^0\implies c=1$

$x=e^{-kt}(1)=e^{-kt}$

<br>

$0.5=e^{-3.3k}$

$k=\frac{\ln(0.5)}{-3.3}=\frac{-0.693}{-3.3}=0.21$

$x=e^{-0.21t}$

<br>

$0.1=e^{-0.21t}$

$\ln(0.1)=-0.21t$

$t=\frac{\ln(0.1)}{-0.21}=\frac{-2.3}{-0.21}=10.96\text{ hours}$

**13.** A thermometer is removed from a room where the temperature is 70째F and is taken outside, where the air 
temperature is 10째F. After one-half minute the thermometer reads 50째F. What is the reading of the thermometer at $t=1\text{ min}$? How long will it take for the thermometer to reach 15째F?

$\frac{dT}{dt}=k(T-T_\text{out})$

$T(0)=70\quad T(0.5)=50$

$\frac{1}{T-T_\text{out}}dT=kdt$

$\int\frac{1}{T-10}dT=k\int1dt$

$\ln(T-10)=kt+c$

$e^{\ln(T-10)}=e^{kt+c}$

$T-10=e^{kt}e^{c}$

$T=e^{kt}e^c+10=ce^{kt}+10$

<br>

$70=ce^0+10$

$c=70-10=60$

$T=60e^{kt}+10$

<br>

$50=60e^{0.5k}+10$

$50-10=60e^{0.5k}$

$0.667=e^{0.5k}$

$\ln(0.667)=0.5k$

$k=\frac{\ln(0.667)}{0.5}=\frac{-0.405}{0.5}=-0.811$

$T=60e^{-0.811t}+10$

$T=60e^{-0.811(1)}+10$

$T=60\cdot\frac{4}{9}+10$

$\frac{80}{3}+10\approx 36.7$

<br>

$15=60e^{-0.811t}+10$

$15-10=60e^{-0.811t}$

$5=60e^{-0.811t}$

$\ln(0.833)=-0.811t$

$t=\frac{\ln(0.833)}{-0.811}=\frac{-2.485}{-0.811}=3.064\text{ minutes}$
