```python
from IPython.display import display, Math
import sympy as sym
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

import sympy.plotting.plot as symplot
from matplotlib.patches import Polygon
from scipy.signal import find_peaks

def sep(): print('-' * 64)
```

```python
# Graphing complex roots of unity
n = 4
for k in range(0,n):
    z = sym.exp(2*sym.pi * sym.I * k/n)
    display(Math('(%s)^{%s} \\Rightarrow %s' %(sym.latex(z), n, sym.latex(z**n))))

for k in range(0,n):
    z = np.exp(2*np.pi * 1j * k/n)
    plt.plot([0, np.real(z)], [0, np.imag(z)])
plt.axis('square')

x = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(x), np.sin(x), color='gray')

plt.show()
```

```python
# Logarithms
a = 3
b = 4

res1 = np.log(a * b)
res2 = np.log(a) * np.log(b)
res3 = np.log(a) + np.log(b)

display(Math('\\log(%g)\\times%g) = %g' %(a, b, res1)))
display(Math('\\log(%g)\\times \\log(%g)) = %g' %(a, b, res2)))
display(Math('\\log(%g) + \\log(%g)) = %g' %(a, b, res3)))
```

```python
# Arithmetic and Geometric Sequences
a = 2
d = 3
maxn = 10

arith_seq = a + d*np.arange(0, maxn)
print(arith_seq)

a = 2
r = 3
geo_seq = a * r**np.arange(0, maxn)
print(geo_seq)

plt.plot(arith_seq, 'ks', label='arithmetic')
plt.plot(geo_seq, 'ro', label='geometric')
plt.legend()

plt.show()
```

```python
# Maxima and Minima of Functions (extrema)
lst = [-4, 5.5, 1, 2]

maxval = np.max(lst)
maxidx = np.argmax(lst)
print(lst[maxidx])

x = np.linspace(0, 2*np.pi, 20)
fx = -(np.cos(x) + np.sqrt(x))

fmax = np.max(fx)
fmaxidx = np.argmax(fx)

plt.plot(x, fx, 'bo-')
plt.plot(x[fmaxidx], fmax, 'rs')

plt.plot(x, fx)
plt.show()


# Min
x = np.linspace(-2, 2, 51)
fx = x**3 + x**4

fminidx = np.argmin(fx)

plt.plot(x, fx, 'bs-')
plt.plot(x[fminidx], fx[fminidx], 'rs')

plt.xlim([-1.2, 0.5])
plt.ylim([-0.2, 0.2])
plt.show()
```

```python
# Even and Odd Functions
x = np.linspace(-5, 5, 20)

feven = x**2
feven_neg = (-x)**2

plt.plot(x, feven)
plt.plot(x, feven_neg, 'ro')

plt.show()

fodd = x**3
fodd_neg = (-x)**3
fneg_odd = -fodd

plt.plot(x, fodd)
plt.plot(x, fodd_neg, 'ro')
plt.plot(x, fneg_odd, 'g')

plt.show()
```

```python
# Pythagorean Theorem
a = 3
b = 4
c = np.sqrt(a**2 + b**2)

plt.plot([0,a], [0,0], 'k', linewidth=2)
plt.plot([0,0], [0,b], 'k', linewidth=2)
plt.plot([0,a], [b,0], 'k', linewidth=2)

plt.plot([0.3, 0.3], [0, 0.3], 'k', linewidth=1)
plt.plot([0, 0.3], [0.3, 0.3], 'k', linewidth=1)

plt.text(a/3, -0.50, f'a={a}', fontsize=12)
plt.text(-1.0, b/3, f'b={b}', fontsize=12)
plt.text(a/3 + 0.50, b/3 + 1.0, f'c={int(c)}', fontsize=12)

plt.axis('square')
axlim = np.max((a,b)) + 0.5
plt.axis([-0.5, axlim, -0.5, axlim])
plt.axis('off')

plt.show()
```

```python
# Piecewise Functions

x = sym.symbols('x')

piece1 = 0
piece2 = -2 * x
piece3 = x**3 / 10

fx = sym.Piecewise((piece1, x < 0), (piece2, (x >= 0) & (x < 4)), (piece3, x >= 4))
fxx = sym.lambdify(x, fx)
xx = np.linspace(-3, 5, 1234)

plt.plot(xx, fxx(xx))

plt.show()
```

```python
# Derivatives of Polynomials
display(Math('\\frac{d}{dx}(ax^n) = nax^{n-1}'))

x = sym.symbols('x')
fx = x**2
dfx = sym.diff(fx)

# Leibniz notation
display(Math('f(x) = %s, \\quad \\frac{df}{dx}=%s' %(sym.latex(fx), sym.latex(dfx))))

# Lagrange notation
display(Math('f(x) = %s, \\quad f\'=%s' %(sym.latex(fx), sym.latex(dfx))))

# Newton notation
display(Math('f(x) = %s, \\quad \\dot{f}=%s' %(sym.latex(fx), sym.latex(dfx))))


# Plotting a derivative
fx = 3 - x**3
p = symplot(fx, (x, -5, 5), show=False)
p.extend(symplot(sym.diff(fx), (x,-5,5), show=False))

p[1].line_color = 'r'
p[0].label = '$f(x) = %s$' %sym.latex(fx)
p[1].label = '$f(x) = %s$' %sym.latex(sym.diff(fx))

p.legend = True
p.xlim = [-3, 3]
p.ylim = [-10, 10]

p.show()
```

```python
# Derivatives of Trig functions
display(Math('\\frac{d}{dx}(\\cos(x)) = -\\sin(x)'))
display(Math('\\frac{d}{dx}(-\\sin(x)) = -\\cos(x)'))
display(Math('\\frac{d}{dx}(-\\cos(x)) = \\sin(x)'))
display(Math('\\frac{d}{dx}(\\sin(x)) = \\cos(x)'))
sep()

q = sym.symbols('q')
print(sym.diff(sym.cos(q)))
print(sym.diff(sym.sin(q)))
```

```python
# Graphing function tangent line
x = sym.symbols('x')

f = x**2
df = sym.diff(f)

xa = 1
fa = f.subs(x, xa)
dfa = df.subs(x, xa)

xx = np.linspace(-2, 2, 200)
f_fun = sym.lambdify(x, f)(xx)
df_fun = sym.lambdify(x, df)(xx)

tanline = dfa * (xx - xa) + fa
plt.plot(xx, f_fun, label='f(x)')
plt.plot(xx, tanline, label='tangent')
plt.plot(xa, fa, 'ro')

plt.axis('square')
plt.axis([-2,2,-0.5,2])

ax = plt.gca()
plt.plot(ax.get_xlim(), [0,0], 'k--')
plt.plot([0,0], ax.get_ylim(), 'k--')
plt.legend()

plt.show()
```

```python
# Critical points of a function

# empirical method
x = np.linspace(-5, 5, 1001)
fx = x**2 * np.exp(-x**2)
dfx = np.diff(fx) / (x[1] - x[0])

localmax = find_peaks(fx)[0]
localmin = find_peaks(-fx)[0]

print('The critical points are ' + str(x[localmax]) + ' ' + str(x[localmin]))

plt.plot(x, fx)
plt.plot(x[0:-1], dfx)

plt.plot(x[localmax], fx[localmax], 'ro')
plt.plot(x[localmin], fx[localmin], 'go')

plt.show()


# symbolic method
x = sym.symbols('x')
fx = x**2 * sym.exp(-x**2)
dfx = sym.diff(fx)

crit_points = sym.solve(dfx)
```

```python
# Partial Derivatives
display(Math('f(x,y) = 2xy^2'))
display(Math('\\frac{\\partial f}{\\partial x} = f_x = 2y^2'))
display(Math('\\frac{\\partial f}{\\partial y} = f_y = 4xy'))
sep()

x,y = sym.symbols('x,y')
f = x**2 + x*y**2
display(Math('\\frac{\\partial f}{\\partial x} = %s' %sym.latex(sym.diff(f, x))))
display(Math('\\frac{\\partial f}{\\partial y} = %s' %sym.latex(sym.diff(f, y))))
```

```python
# Indefinite and definite integrals

display(Math('\\int_a^b f(x)\:dx'))                        # definite integrals
display(Math('\\int bx^a\:dx = \\frac{bx^{a+1}}{a+1}+c'))  # indefinite integral
sep()

x = sym.symbols('x')
f = x

print(sym.integrate(f)) # indefinite

p = sym.plotting.plot(f, show=False)
p.xlim = [0,1]
p.ylim = [0,1]
p.show()

print(sym.integrate(f,(x,0,1))) # definite


f = x**3 / (x-2)
intf = sym.integrate(f)

p = sym.plotting.plot(f, show=False)
p.extend(sym.plotting.plot(intf, (x,2.1,10), show=False, line_color='r'))
p.ylim = [-200,200]
p[0].label = '$f(x) = %s$' %(sym.latex(f))
p[1].label = '$\\int f(x) dx = %s$' %(sym.latex(intf))
p.legend = True

p.show()
```

```python
# Area Between Two Curves
x = sym.symbols('x')
symf = x**2
symg = x
f = sym.lambdify(x, symf)
g = sym.lambdify(x, symg)

xx = np.linspace(-2, 2, 55)
xpatch = np.linspace(0, 1, 100)
ypatch = np.vstack((g(xpatch), f(xpatch))).T

fig,ax = plt.subplots()
ax.add_patch(Polygon(ypatch, facecolor='k', alpha=0.3))

plt.plot(xx, f(xx))
plt.plot(xx, g(xx), 'r')
plt.axis([-0.25, 1.25, -0.5, 1.5])
plt.legend(['$f(x)=%s$' %sym.latex(symf), '$g(x)=%s$' %sym.latex(symg)])

plt.show()
```
