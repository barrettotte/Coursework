**3.** Evaluate the line integral $\int_C yz\cos(x)ds$, $C: x=t,\;y=3\cos(t),\;z=3\sin(t),\; 0\le t\le\pi$

$ds=\sqrt{(\frac{dx}{dt})^2+(\frac{dy}{dt})^2+(\frac{dz}{dt})^2}$

$ds=\sqrt{(1)^2+(-3\sin(t))^2+(3\cos(t))^2}dt$

$=\sqrt{1+9\sin^2(t)+9\cos^2(t)}dt$
$=\sqrt{1+9}dt$
$=\sqrt{10}dt$

<br>

$\int_C yz\cos(x)ds$

$=\int_0^\pi (3\cos(t))(3\sin(t))\cos(t)\sqrt{10}dt$

$=9\sqrt{10}\int_0^\pi \cos^2(t)\sin(t)dt$

<br>

let $u=\cos(t)\implies\frac{du}{dt}=-\sin(t)$

$du=-sin(t)dt\implies dt=(-\frac{1}{\sin(t)})du$

$\int u\sin(t)u(-\frac{1}{\sin(t)})du$
$=\int -u^2du$

adjust boundaries:

$t=0 \implies u=\cos(0)=1$

$t=\pi \implies u=\cos(\pi)=-1$

<br>

$9\sqrt{10}\int_1^{-1}-u^2du$
$=9\sqrt{10}\big[\frac{u^3}{3}\big]_{-1}^1$

$9\sqrt{10}(\frac{1}{3}-(-\frac{1}{3}))$
$=9\sqrt{10}\frac{2}{3}$
$=6\sqrt{10}$

**answer:** $6\sqrt{10}$

**11.** Show that $\vec{F}$ is a conservative vector field and 
find a function $f$ such that $\vec{F}=\nabla f$.
$\quad\vec{F}(x,y)=(1+xy)e^{xy}\hat{i}+(e^y+x^2e^{xy})\hat{j}$

$\vec{F}=P\hat{i}+Q\hat{j}$ is conservative if $P_y=Q_x$

$P=(1+xy)e^{xy}$

$P_y=\frac{\partial}{\partial y}((1+xy)e^{xy})$

$\frac{\partial}{\partial y}(1+xy)e^{xy}+\frac{\partial}{\partial y}(e^{xy})(1+xy)$

$=xe^{xy}+e^{xy}x(1+xy)$
$=(2+xy)xe^{xy}$

<br>

$Q=e^y+x^2e^{xy}$

$Q_y=\frac{\partial}{\partial x}(e^y+x^2e^{xy})$

$=\frac{\partial}{\partial x}(e^y)+\frac{\partial}{\partial x}(x^2e^{xy})$

$=0+2xe^{xy}+e^{xy}yx^2$
$=(2+xy)xe^{xy}$

<br>

**answer:** $P_y=Q_x$ so $\vec{F}$ is conservative

**17.** Use Green's theorem to evaluate $\int_C x^2ydx-xy^2dy$, where $C$
is the circle $x^2+y^2=4$ with counterclockwise orientation

$\int_C Pdx + Qdy = \iint\limits_D\left(\frac{\partial Q}{\partial x}-\frac{\partial P}{\partial y}\right)dA$

$\int_C x^2ydx-xy^2dy=\iint\limits_D\left( \frac{\partial(-xy^2)}{\partial x}-\frac{\partial(x^2y)}{\partial y} \right)dA$

$=\iint\limits_D -y^2-x^2 dA$

<br>

Transformation:

$D: x^2+y^2=4$

$\{(r,\theta)\in D \;|\; 0\le r\le 2,\; 0\le\theta\le 2\pi \}$

$x=r\cos\theta\quad y=r\sin\theta$

$\iint\limits_D-(r\cos\theta)^2-(r\sin\theta)^2 \cdot\frac{\partial(x,y)}{\partial(r,\theta)} dA$

<br>
Jacobian:

$x=f(r,\theta)=r\cos(\theta)\quad y=g(r,\theta)=r\sin(\theta)$

$f_r=\cos(\theta),\quad f_\theta=-r\sin(\theta)$

$g_r=\sin(\theta),\quad g_\theta=-r\cos(\theta)$

$J=|(\cos(\theta))(-r\cos(\theta))-(-r\sin(\theta))(\sin(\theta))| = r$

<br>

$\int_0^2\int_0^{2\pi}-r^2(r)d\theta dr$

$\int_0^{2\pi} -r^3 dr$
$=\big[-\frac{r^4}{4}\big]_0^{2\pi}$
$=-4\pi^4$

$\int_0^2 -4\pi^4 dr$
$=\big[-4\pi^4r\big]_0^2$
$=-8\pi^4$

**answer:** $-8\pi^4$

**25.** Find the area of the part of the surface $z=x^2+2y$ that lies above
the triangle with vertices $(0,0),\;(1,0),\;(1,2)$

$\iint\limits_S dS=\iint\limits_D\sqrt{1+(\frac{\partial z}{\partial x})^2+(\frac{\partial z}{\partial y})^2}dA$

For $(0,0)$ and $(1,2)$:
$\frac{y-0}{x-0}=\frac{2-0}{1-0}$
$\implies y=2x$

$\{(x,y)\in D\;|\;0\le y\le 2x,\;0\le x\le 1\}$

<br>

$\int_0^1\int_0^{2x}\sqrt{1+(2x)^2+(2)^2}\;dydx$

$\int_0^1\int_0^{2x}\sqrt{5+4x^2}dydx$
$=\int_0^1\big[y\sqrt{5+4x^2}\big]_0^{2x}dx$

$=\int_0^1 2x\sqrt{5+4x^2}dx$
$=2\int_0^1 x\sqrt{5+4x^2}dx$

let $u=5+4x^2 \implies \frac{du}{dx}=8x$

$du=8xdx \quad dx=\frac{1}{8x}du$

$\int x\sqrt{u}\frac{1}{8x}du$
$=\int\frac{\sqrt{u}}{8}du$

adjust boundaries: 

$x=0 \implies u=5+4(0)^2 = 5$

$x=1 \implies u=5+4(1)^2 = 9$

$2\int_5^9 \frac{\sqrt{u}}{8}du$
$=\frac{1}{4}\int_5^9 \sqrt{u}du$
$=\frac{1}{4}\big[\frac{u^{3/2}}{3/2}\big]_5^9$
$=\frac{1}{4}\big[\frac{2u^{3/2}}{3}\big]_5^9$

$=\frac{1}{4}(18-\frac{10\sqrt{5}}{3})$
$=\frac{27-5\sqrt{5}}{6}$

**answer:** $\frac{27-5\sqrt{5}}{6}$

**27.** Evaluate the surface integral $\iint\limits_S zdS$, where $S$ is the part of the
paraboloid $z=x^2+y^2$ that lies under the plan $z=4$

$dS=\sqrt{1+(\frac{\partial z}{\partial x})^2+(\frac{\partial z}{\partial y})^2}dydx$

$z=x^2+y^2 \implies dS=\sqrt{1+(2x)^2+(2y)^2}dydx$

$dS=\sqrt{1+4x^2+4y^2}dydx$

<br>

$\iint\limits_S z dS=\iint\limits_D z\sqrt{1+4x^2+4y^2}dxdy$

$=\iint\limits_D(x^2+y^2)\sqrt{1+4x^2+4y^2}dxdy$

$S$ is part of paraboloid below plane $z=4\implies$ 
$D$ is region inside circle $x^2+y^2=4$

<br>

To polar:

$\int_0^{2\pi}\int_0^2 r^2\sqrt{1+4r^2}\;rdrd\theta$

$\int_0^{2\pi}d\theta=2\pi$

$\int_0^2 r^2\sqrt{1+4r^2}rdr$

<br>

let $u=r^2 \implies \frac{du}{dr}=2r,\quad du=2rdr,\quad dr=\frac{1}{2r}du$

$\int ur\sqrt{1+4u}\frac{1}{2r}du$
$=\int \frac{u\sqrt{1+4u}}{2}du$

adjust boundaries:

$r=0 \implies u=(0)^2=0$

$r=1 \implies u=(1)^2=1$

$\frac{1}{2}\int_0^1u\sqrt{1+4u}du$

let $v=1+4u \implies \frac{dv}{du}=4,\quad dv=4du,\quad du=\frac{1}{4}dv$

$\int u\sqrt{v}\frac{1}{4}dv$
$=\int \frac{\sqrt{v}(v-1)}{16}dv$

adjust boundaries:

$u=0 \implies v=1+4(0)=1$

$u=1 \implies v=1+4(1)=5$

$\frac{1}{2}\int_1^5 \frac{\sqrt{v}(v-1)}{16}dv$
$=\frac{1}{32}\sqrt{v}(v-1)dv$

$\frac{1}{32}\left( \int_1^5v^{3/2}dv - \int_1^5\sqrt{v}dv \right)$

$\frac{1}{32}\left( (10\sqrt{5}-\frac{2}{5})-(\frac{10\sqrt{5}}{3}-\frac{2}{3}) \right)$

$=\frac{25\sqrt{5}+1}{120}$

<br>

$2\pi\cdot \frac{25\sqrt{5}+1}{120}$

**answer:** $\frac{\pi 25\sqrt{5}+1}{60}$
