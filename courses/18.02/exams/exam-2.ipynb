**1.** Let $f(x,y)=xy-x^4$

a). Find the gradient of $f$ at $P: (1,1)$

$\nabla f = \langle f_x, f_y\rangle$

$f_x =\frac{\partial}{\partial x}(xy-x^4)$
$=y-4x^3$

$f_y =\frac{\partial}{\partial y}(xy-x^4)$
$=x$

$\nabla f = \langle y-4x^3,\; x \rangle$

at $P \nabla f(1,1) = \langle -3, 1 \rangle$

**answer:** $\langle -3,1 \rangle$

<br>

b). Give an approximate formula telling how small changes $\Delta x$ and $\Delta y$ produce a small change
$\Delta W$ in the value of $w = f(x,y)$ at the point $(x,y) = (1,1)$

**answer:** $\Delta w \approxeq -3\Delta x + \Delta y$

**3.** Find the equation of the tangent plane to the surface $x^3y+z^2=3$ at the point $(-1,1,2)$

$\nabla f=\langle f_x, f_y, f_z \rangle$

$f_x = \frac{\partial}{\partial x}(x^3y+z^2) = 3x^2 y$

$f_y = \frac{\partial}{\partial y}(x^3y+z^2) = x^3$

$f_z = \frac{\partial}{\partial z}(x^3y+z^2) = 2z$

$\nabla f \langle 3x^2y,\; x^3,\; 2z \rangle$

<br>

$\nabla f(-1,1,2) = \langle 3(-1)^{2(1)},\; (-1)^3,\; 2(2) \rangle$

$\nabla f(-1,1,2) = \langle 3,-1,4 \rangle$

<br>

$0 = f_x(x_0,y_0,z_0)(x-x_0) + f_y(x_0,y_0,z_0)(y-y_0) +f_z(x_0,y_0,z_0)(z-z_0)$

$0 = 3(x-(-1)) + -1(y-1) + 4(z-2)$

$0 = 3x+4 -y + 4z-8$

**answer:** $3x-y+4z=4$

**5.** In problem 4 above, instead of substituting for $z$, one could also use Lagrange multipliers to maximize the volume $V=xyz$ with the same constraint $x^2+y^2+z=1$.

a). Write down the Lagrange multiplier equations for this problem.

$f(x,y,z)=xyz$

$g(x,y,z) = x^2+y^2+z-1$

<br>

$\nabla f(x_0,y_0,z_0) = \lambda \nabla g(x_0,y_0,z_0)$

$yz = 2x\lambda \implies xyz = 2x^2\lambda$
$\implies \frac{xyz}{\lambda}=2x^2$

$xz = 2y\lambda \implies xyz = 2y^2\lambda$
$\implies \frac{xyz}{\lambda}=2y^2$

$xy = \lambda \implies xyz = z\lambda$
$\implies \frac{xyz}{\lambda}=z$

$2x^2=2y^2=z$

<br>

b). Solve the equations (still assuming $x>0, y>0$)


using above:

$z = 2x^2$
$\implies 4x^2=1$

$x=\frac{1}{2},\quad \frac{1}{2},\quad z=\frac{1}{2}$

**6.** Let $w=f(u,v)$, where $u=xy$ and $v=x/y$. Using the chain rule, express $\frac{\partial w}{\partial x}$ and $\frac{\partial w}{\partial y}$ in terms of $x,y,f_u,$ and $f_v$

$\frac{\partial w}{\partial x}$
$=f_u u_x + f_v v_x$
$=yf_u + \frac{1}{y}f_v$

$\frac{\partial w}{\partial y}$
$=f_u u_y + f_v v_y$
$=xf_u - \frac{x}{y^2}f_v$
