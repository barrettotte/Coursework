**1.** A unit cube lines in the first octant, with a vertex at the origin (see figure).

![assets/exam-1-01.png](assets/exam-1-01.png)

a). Express the vectors $\overrightarrow{OQ}$ (a diagonal of the cube) and $\overrightarrow{OR}$ 
(joining $O$ to the center of a face) in terms of $\hat{i},\hat{j},\hat{k}$.

$\overrightarrow{OQ} = \hat{i}+\hat{j}+\hat{k}$

$\overrightarrow{OR} = \frac{1}{2}\hat{i}+\hat{j}+\frac{1}{2}\hat{k}$

<br>

b). Find the cosine of the angle between $OQ$ and $OR$

$\cos\theta = \frac{a\cdot b}{|a||b|}$

$|\overrightarrow{OQ}| = \sqrt{3} \quad |\overrightarrow{OR}|=\sqrt{\frac{3}{2}}$

$\cos\theta = \frac{\langle 1,1,1 \rangle \cdot \langle \frac{1}{2},1,\frac{1}{2} \rangle}{\sqrt{3}\sqrt{\frac{3}{2}}}$
$\;=\; \frac{\frac{1}{2}+1+\frac{1}{2}}{\frac{3\sqrt{2}}{2}}$

**answer:** $\frac{2\sqrt{2}}{3}$

$\overrightarrow{R} = 3\cos t\hat{i} +3\sin t \hat{j} + t\hat{k}$.

$v(t) = r'(t) = \langle -3\sin t,\; 3\cos t,\; 1 \rangle$

$s(t) = |v(t)| = \sqrt{(-3\sin t)^2+(3\cos t)^2 +1^2}$

$= \sqrt{9\sin^2t + 9\cos^2t + 1}$
$\;=\; \sqrt{9+1}$

**answer:** 

$v(t) = \langle -3\sin t,\; 3\cos t,\; 1 \rangle$

$s(t) = \sqrt{10}$

**4.** The top extremity of a ladder of length $L$ rests against a vertical wall, while the bottom is being pulled away. Find parametric equations for the midpoint $P$ of the ladder, using as parameter the angle $\theta$ between the ladder and the horizontal ground.

![assets/exam-1-04.png](assets/exam-1-04.png)

Origin to top: 

$\sin\theta = \frac{\text{opposite}}{\text{hypotenuse}}$
$\implies \text{opposite}=\text{hypotenuse}\cdot\sin\theta$

$\overrightarrow{OA} = \langle 0,\; L\sin\theta \rangle$

<br>

Origin to bottom:

$\cos\theta = \frac{\text{adjacent}}{\text{hypotenuse}}$
$\implies \text{adjacent}=\text{hypotenuse}\cdot\cos\theta$

$\overrightarrow{OB} = \langle -L\cos\theta,\; 0 \rangle$

Midpoint: $\overrightarrow{OP} = \frac{1}{2}(\overrightarrow{OA}+\overrightarrow{OB})$
$= \langle \frac{-L}{2}\cos\theta,\; \frac{L}{2}\sin\theta \rangle$

**answer:** $x = \frac{-L}{2}\cos\theta,\quad y=\frac{L}{2}\sin\theta$

**5.**

a) Find the area of the space triangle with vertices $P_0:(2,1,0),\; P_1:(1,0,1),\; P_2:(2,-1,1)$.

$P_0P_1 = (-1,-1,1)$
$\quad P_0P_2 = (0,-2,1)$

$P_0P_1 \times P_0P_2$
$= \begin{vmatrix} \hat{i}&\hat{j}&\hat{k}\\ -1&-1&1\\ 0&-2&1 \end{vmatrix}$
$= \langle 1,1,2 \rangle$

$\text{area} = \frac{1}{2}|P_0P_1\times P_0P_2|$

**answer:** $\frac{1}{2}\sqrt{6}$

<br>

b) Find the equation of the plane containing the three points $P_0,P_1,P_2$.

$a(x-x_0)+b(y-y_0)+c(z-z_0)=0$

$1(x-x_0)+1(y-y_0)+2(z-z_0)=0$

using $P_2: \; 1(x-2)+1(y+1)+2(z-1)=0$

$x-2+y+1+2z-2 = 0$

**answer:** $x+y+2z = 3$

c) Find the intersection of this plane with the line parallel to the vector 
$\overrightarrow{v} \langle 1,1,1 \rangle$ and passing through the point $S: (-1,0,0)$.

$x= x_0+t,\quad y=y_0+t,\quad z=z_0+t$

$x=-1+t,\quad y=t,\quad z=t$

$x+y+2z=3 \Rightarrow -1+t+t+2t=3$

$-1+4t=3 \implies t=1$

$x=-1+1,\quad y=1,\quad z=1$

**answer:** $(0,1,1)$
