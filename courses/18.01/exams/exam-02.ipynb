1. Estimate the following to two decimal places

**a.** $\; \sin\left(\pi+\frac{1}{100}\right)$

let $y = \sin(x)$

$dy = \cos(x)dx$

if $x = \pi$, then $dx = \pi - \pi+\frac{1}{100} = \frac{1}{100}$

$dy = \cos(\pi)\frac{1}{100} = -\frac{1}{100}$

<br>

$\sin(\pi + \frac{1}{100}) = y + dy$

$\sin(\pi + \frac{1}{100}) = \sin(\pi) + -\frac{1}{100} = 0 - \frac{1}{100}$

**answer:** $\quad -0.01$

<br>

**b.** $\; \sqrt{101}$

let $y=\sqrt{x}$

$dy = \frac{1}{2\sqrt{x}}dx \;=\; \frac{dx}{2\sqrt{x}}$

if $x = 100$, then $dx = 101-100 = 1$

$dy = \frac{1}{2\sqrt{100}}(1) = \frac{1}{2\sqrt{100}} = \frac{1}{20} = 0.05$

<br>

$\sqrt{101} = y + dy$

$\sqrt{101} = \sqrt{100} + 0.05$

**answer:** $\quad \sqrt{101} = 10.05$

2. An architect plans to build a triangular enclosure with a fence on two sides and a wall on the third side. Each of the fence segments has fixed length $L$. What is the length $x$ of the third side if the region enclosed has the largest possible area?



<br>

$A = \frac{1}{2}xh \quad L^2 = h^2+(\frac{x}{2})^2 \quad 0 \le x \le 2L$

$\frac{d}{dx}(h^2) + \frac{d}{dx}(\frac{x}{2}^2) = 0$ 

$\frac{d}{dx}(\frac{x}{2}^2)$
$\;=\; \frac{x}{2}$

$2h\frac{d}{dx}(h) + \frac{x}{2} = 0$

$\frac{d}{dx}(h) = -\frac{x}{4h}$

$A' = \frac{1}{2}(h+xh)$
$\;=\; \frac{1}{2}(h + x(-\frac{x}{4h}))$
$\;=\; \frac{x^2}{4h}$

$L^2 = (\frac{x^2}{4h})^2 + \frac{x^2}{4}$
$\quad\implies x^2 = 2L^2 $

**answer:** $\quad x = L\sqrt{2}$

3. A rocket has launched straight up, and its altitude is $h=10t^2$ feet after $t$ seconds. You are on the ground 1000 feet from the launch site. The line of sight from you to the rocket makes an angle $\theta$ with the horizontal. By how many Radians per second is $\theta$ changing ten seconds after the launch?



<br>

$\tan(\theta) = \frac{h}{1000}$
$\;=\; \frac{10t^2}{1000} \;=\; \frac{1}{100}t^2$

$\sec^2(\theta)\frac{d\theta}{dt} \;=\; 2\frac{1}{100}t$
$\;=\; \frac{1}{50}t$

$\tan(\theta) = 1$

$\sec^2(\frac{\pi}{4})\frac{d\theta}{dt}$
$\;=\; \frac{1}{50}(10) \;=\; \frac{1}{5}$

$\frac{1}{5}\cos^2(\frac{\pi}{4})$
$\;=\; \frac{1}{5}\left(\frac{\sqrt{2}}{2}\right)^2$
$\;=\; \frac{1}{10}$

**answer:** $\quad \frac{1}{10}$ rad/s

4. Evaluate the following indefinite integrals

**a** $\; \int\cos(3x)dx$

let $u = 3x$

$\frac{du}{dx} = 3 \implies du=3dx$

$dx = \frac{1}{3}du$

$\int\cos(u)\frac{1}{3}du$
$\;=\; \frac{1}{3}\int\cos(u)du$
$\;=\; \frac{1}{3}\sin(u)$

**answer:** $\quad \frac{1}{3}\sin(3x) + c$

<br>

**b.** $\; \int xe^{x^2}dx$

let $u=x^2$

$\frac{du}{dx} = 2x \implies du = 2xdx \implies dx = \frac{1}{2x}du$

$\int xe^u\frac{1}{2x}du$
$\;=\; \int\frac{e^u}{2}du$
$\;=\; \frac{1}{2}\int e^u du$

$= \frac{1}{2}e^u$

**answer:** $\quad \frac{1}{2}e^{x^2} + c$
