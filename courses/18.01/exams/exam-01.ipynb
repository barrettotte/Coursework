1. Find the tangent line to $y = \frac{1}{3}x^2$ at $x=1$

$\frac{d}{dx}\left(\frac{1}{3}x^2\right) \;=\; \frac{2}{3}x = f'(x)$

$f'(1) = \frac{2}{3}$

$y_0 = f(1) = \frac{1}{3}$

$y-y_0 = f'(1)(x-x_0) \quad \to \quad y-\frac{1}{3} = \frac{2}{3}(x-1)$

**answer:** $\quad\quad y = \frac{2}{3}x - \frac{1}{3}$

2. Find the derivative of the following functions:

$\text{a.}\;\;\frac{x}{\sqrt{1-x}}$

$\frac{d}{dx}\left(\frac{x}{\sqrt{1-x}}\right)$
$\;=\; \frac{\frac{d}{dx}(x)\sqrt{1-x}-\frac{d}{dx}(\sqrt{1-x})x}{(\sqrt{1-x})^2} \quad\text{(Quotient Rule)}$

$\frac{d}{dx}(x) = 1$

$\frac{d}{dx}(\sqrt{1-x})$

$\quad \text{let } f=\sqrt{u},\;u=1-x\quad\text{(Chain Rule)}$

$\quad \frac{d}{du}(\sqrt{u})\frac{d}{dx}(1-x)$

$\quad = \frac{1}{2\sqrt{u}}\frac{d}{dx}(1-x)$
$\;=\; \frac{1}{2\sqrt{1-x}}(-1) \;=\; \frac{1}{2\sqrt{1-x}}$

$= \frac{1 \cdot \sqrt{1-x} - \left(-\frac{1}{2\sqrt{1-x}}\right)x}{(\sqrt{1-x})^2}$
$\;=\; \frac{\sqrt{1-x}+\frac{1}{2\sqrt{1-x}}}{1-x} $
$\;=\; \frac{\frac{2-x}{2\sqrt{1-x}}}{1-x}$

**answer:** $\quad \frac{2-x}{2\sqrt{1-x}(1-x)}$

<br>


$\text{b.}\;\;\frac{\cos(2x)}{x}$

$\frac{d}{dx}(\frac{\cos(2x)}{x}) = \frac{\frac{d}{dx}(\cos(2x))x - \frac{d}{dx}(x)\cos(2x)}{x^2}\quad \text{(Quotient Rule)}$

$\frac{d}{dx}(x) = 1$

$\frac{d}{dx}(\cos(2x))$

$\quad\text{let }f=\cos(u),\;u=2x\quad\text{(Chain Rule)}$

$\quad\frac{d}{du}(\cos(u))\frac{d}{dx}(2x)$
$\;=\; -\sin(u)\cdot 2$
$\;=\; -2\sin(2x)$

$= \frac{(-2\sin(2x))x - \cos(2x)}{x^2}$

**answer:** $\quad \frac{-2x\sin(2x)-\cos(2x)}{x^2}$

<br>

$\text{c.}\;\;e^{2f(x)} = g(x)$

**answer:** $\frac{d}{dx}(g(x)) = 2\frac{d}{dx}(f(x)) e^{2f(x)}$

<br>

$\text{d.}\;\;\ln(\sin(x))$

$\frac{d}{dx}(\ln(\sin(x)))$

$\text{let } f=\ln(u),\;u=\sin(x)\quad\text{(Chain Rule)}$

$\frac{d}{du}(\ln(u))\frac{d}{dx}(\sin(x))$
$\;=\; \frac{1}{u}\frac{d}{dx}(\sin(x))$
$\;=\; \frac{1}{\sin(x)}\frac{d}{dx}(\sin(x))$

$\frac{d}{dx}(sin(x)) = \cos(x)$

$= \frac{1}{\sin(x)}\cos(x) \;=\; \frac{\cos(x)}{\sin(x)} \;=\; \cot(x)$

**answer:** $\quad \cot(x)$

3. Find $\frac{dy}{dx}$ for the function $y$ defined implicitly by $y^4 + xy = 4 \text{ at } x=3,\;y=1$

$\frac{d}{dx}(y^4+xy) = \frac{d}{dx}(4)$

$\frac{d}{dx}(y^4+xy) = \frac{d}{dx}(y^4)+\frac{d}{dx}(xy)$

$\frac{d}{dx}(y^4)$
$\;=\; 4y^3\frac{d}{dx}(y)$

$\frac{d}{dx}(xy)$
$\;=\; \frac{d}{dx}(x)y + \frac{d}{dx}(y)x$
$\;=\; 1y + x\frac{d}{dx}(y)$

<br>

$4y^3\frac{d}{dx}(y) + y + x\frac{d}{dx}(y) = 0$

$4y^3\frac{d}{dx}(y) + x\frac{d}{dx}(y) = -y$

$\frac{d}{dx}(y)(4y^3+x)) = -y$

$\frac{d}{dx}(y) = -\frac{y}{4y^3+x}$

<br>

$\frac{d}{dx}(y) \;=\; \frac{-(1)}{4(1)^3 + (3)} \;=\; -\frac{1}{7}$

**answer:** $\quad-\frac{1}{7}$

\begin{cases}
ax+b \quad x\lt 1 \\
x^4+x+1 \quad x\ge 1 
\end{cases}$

Find all $a$ and $b$ such that the function $f(x)$ is differentiable.


$\lim_{x \to 1^-} (ax + b) \;=\; a+b$

$\lim_{x \to 1^+} (x^4+x+1) \;=\; 3$

Continuous at $x=1$ when $a+b=3$
<br>

$\frac{d}{dx}(x^4+x+1) \;=\; 4x^3+1$

$\lim_{x \to 1^-} (a) = a$

$\lim_{x \to 1^+}(4x^3+1) = 5$

$a + b = 3 \quad\to\quad (5)+b=3$

**answer:** $\quad a=5,\;b=-2$

5. Derive the formula $\frac{d}{dx}a^x = M(a)a^x$ directly from the definition of the derivative, and identify M(a) as a limit.

$\frac{d}{dx}(a^x) = \lim_{b\to 0}\frac{a^{x+b}-a^x}{b}$
$\;=\;\lim_{b\to 0}\frac{a^xa^b-a^x}{b}$

$=\lim_{b\to 0}\frac{a^x(a^b-1)}{b}$
$\;=\;a^x\lim_{b\to 0}\frac{a^b-1}{b} \equiv M(a)$
