**(1)** $\quad \lim_{x \to 5} \frac{x^2-25}{x^2-6x+5}$

$\quad =\frac{(x+5)(x-5)}{x^2 - 6x + 5}$

$\quad =\frac{(x+5)(x-5)}{(x-1)(x-5)}$

$\quad =\frac{(x+5)}{(x-1)} =\frac{(5)+5}{(5)-1} = \frac{5}{2}$
