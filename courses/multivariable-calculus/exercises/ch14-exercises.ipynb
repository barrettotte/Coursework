**13.** Find the first partial derivative. $f(x,y)=(5y^3+2x^2y)^8$

Solving $F_x$

$\frac{df(u)}{dx}=\frac{df}{du}\cdot\frac{du}{dx}$ (chain rule)

let $u=(5y^3+2x^2y)$

$\frac{\partial}{\partial u}(u^8)\frac{\partial}{\partial x}(5x^3+2x^2y)$
$=8u^7\frac{\partial}{\partial x}(5y^3+2x^2y)$

$8(5y^3+2x^2y)^7\frac{\partial}{\partial x}(5y^3+2x^2y)$
$=8(5y^3+2x^2y)^7\cdot 4yx$

**answer:** $F_x = 32yx(5y^3+2yx^2)^7$

<br>

Solving $F_y$

let $u=(5y^3+2x^2y)$
$\frac{\partial}{\partial u}(u^8)\frac{\partial}{\partial y}(5x^3+2x^2y)$
$=8u^7\frac{\partial}{\partial y}(5y^3+2x^2y)$

$8(5y^3+2x^2y)^7\frac{\partial}{\partial y}(5y^3+2x^2y)$

**answer:** $F_y = 8(5y^3+2x^2y)^7(15y^2+2x^2)$

**15.** Find the first partial derivative. $F(\alpha,\beta)=\alpha^2\ln(\alpha^2+\beta^2)$

Solving $F_\alpha$

$(f\cdot g)'=f'\cdot g+f\cdot g'$ (product rule)

$\frac{\partial}{\partial\alpha}(\alpha^2)\ln(\alpha^2+\beta^2)$
$+ \frac{\partial}{\partial\alpha}(\ln(\alpha^2+\beta^2))\alpha^2$

**answer:** $F_\alpha=2\alpha\ln(\alpha^2+\beta^2)+\frac{2\alpha^3}{\alpha^2+\beta^2}$

<br>

Solving $F_\beta$

$\alpha^2\frac{\partial}{\partial\beta}(\ln(\alpha^2+\beta^2))$

$\alpha^2\frac{\partial}{\partial u}(\ln(u))\frac{\partial}{\partial \beta}(\alpha^2+\beta^2)$
$=\frac{1}{u}\frac{\partial}{\partial\beta}(\alpha^2+\beta^2)$

$\alpha^2\frac{1}{\alpha^2+\beta^2}\cdot 2\beta$

**answer:** $F_\beta=\frac{2\alpha^2\beta}{\alpha^2+\beta^2}$

**25.** Find equation of the tangent line and the normal line. $z=3x^2-y^2+2x,\; (1,-2,1)$

$z-z_0=\frac{\partial z}{\partial x}(x-x_0)+\frac{\partial z}{\partial y}(y-y_0)$

Solving for $\frac{\partial z}{\partial x}$

$\frac{\partial}{\partial x}(2x^2)-\frac{\partial}{\partial x}(y^2)$
$+\frac{\partial}{\partial x}(2x)$
$=6x-0+2$

$F_x=6x+2$

at $(1,-2,1)$: $\;6(1)+2=8$


<br>

Solving for $\frac{\partial z}{\partial y}$

$\frac{\partial}{\partial y}(2x^2)-\frac{\partial}{\partial y}(y^2)$
$+\frac{\partial}{\partial y}(2x)$
$=0-2y+0$

$F_y=-2y$

at $(1,-2,1)$: $\;-2(-2)=4$

<br>

$z-1 = \frac{\partial z}{\partial x}(x-1)+\frac{\partial z}{\partial y}(y+2)$

$z-1 = 8(x-1)+4(y+2)$
$= 8x+4y$

**answer:** $z = 8x+4y+1$

<br>

$8x+4y-z=-1 \implies \langle 8,4,-1 \rangle$

Normal line passes through $(1,-2,1)$ and has normal vector $\langle 8,4,-1 \rangle$

**answer:** $\frac{x-1}{8}=\frac{y+2}{4}=\frac{z-1}{-1}$

**43.** Find the gradient of the function $f(x,y,z)=x^2e^{yz^2}$


$\nabla f = \langle f_x,f_y,f_z \rangle$

solving for $f_x$

$e^{yz^2}\cdot\frac{\partial}{\partial x}(x^2)$
$= e^{yz^2}\cdot 2x$

$f_x = 2e^{yz^2}x$

<br>

solving for $f_y$

$e^{yz^2}\frac{\partial}{\partial y}(yz^2)$
$= e^{yz^2}z^2$

$f_y = e^{yz^2}z^2$

<br>

solving for $f_z$

$x^2\frac{\partial}{\partial z}(e^{yz^2})$

$\frac{df(u)}{dx}=\frac{df}{du}\cdot\frac{du}{dx}$ (chain rule)

$\frac{\partial}{\partial u}(e^u)\frac{\partial}{\partial z}(yz^2)$
$=e^u\frac{\partial}{\partial z}(yz^2)$

$f_z = x^2 e^{yz^2}\cdot 2yz$

<br>

**answer:** $\langle 2e^{yz^2}x,\; e^{yz^2}z^2,\; x^2 e^{yz^2}\cdot 2yz \rangle$

**45.** Find the directional derivative of $f$ at the given point in the indicated direction.\
$f(x,y)=x^2e^{-y}$, $(-2,0)$ in the direction toward point $(2,-3)$

$D_u f(x_0,y_0) = \langle f_x(x_0,y_0),\; f_y(x_0,y_0)\rangle$
$\cdot\; \langle a,b \rangle$

$f_x=$
$e^{-y}\frac{\partial}{\partial x}(x^2)$
$=2e^{-y}x$

$f_y=$
$x^2\frac{\partial}{\partial y}(e^{-y})$
$=e^{-y}\frac{\partial}{\partial}(-y)$
$=x^2e^{-1}(-1)$
$=-e^{-y}x^2$

<br>

$f_x(-2,0) = 2(-2)e^0 = -4$

$f_y(-2,0) = -(-2)^2e^0 = -4$

<br>

$\overrightarrow{PS} = \langle 2-(-2), -3-0\rangle$
$=\langle 4,-3 \rangle$

$u = \frac{\langle 4,-3 \rangle}{|\langle 4, -3 \rangle|}$
$=\frac{\langle 4,-3 \rangle}{\sqrt{(4)^2+(-3)^2}}$
$=\langle \frac{4}{5}, -\frac{3}{5} \rangle$

<br>

$D_u(-2,0) = \frac{4}{5}(-4)-\frac{3}{5}(-4)$

**answer:** $-\frac{4}{5}$
